```python
n, m = map(int, input().split())
ab_list = [list(map(int, input().split())) for _ in range(m)]
result = [[] for _ in range(n)]

for elem in ab_list:
  a = elem[0]
  b = elem[1]
  #from a to b
  result[a-1].append(b)
  #from b to a
  result[b-1].append(a)

for i, elem in enumerate(result):
  length = len(result[i])
  result[i].sort()
  result[i].insert(0, length)
  line = [str(a) for a in result[i]]
  line = ' '.join(line)
  print(line)
```
